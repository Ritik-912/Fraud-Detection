multi-colinearity

```python
# converting categorical features to numerical
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data = df.copy()
for c in data.select_dtypes(include='object'):
    data[c] = encoder.fit_transform(data[c].astype(str))
print(data.info())
print(data.head(4))
```

We convert the categorical features to numerical for applying correlation test on them.

And variance inflation factor is a value that determine the correlation between variables.
If the value is high near about and greater they are highly vorrelated with other variables.

```python
# Calculating Variance Inflation factor for comparing multi-collinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
import seaborn as sns

def mcolinearTest(Df):
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(Df.values, i) for i in range(Df.shape[1])]
    vif["columns"] = data.columns
    print(vif)

    # visualizing through a correlation Heatmap
    correlation = Df.corr()
    plt.figure(figsize=(22,11))
    sns.heatmap(correlation , annot=True)
```

```python
mcolinearTest(data)
```

Feature engineering:
We get to know that oldbalanceOrg and newbalanceOrig are highly correlated
Similarly oldbalanceDest and newbalanceDest and nameOrig and nameDest.
We combine the features by mathematical sum and subtraction.
For the step column we know that they are hours value so we can get idea about the time in a week when fraudulent transactions are significant, for that we convert them in sine, cosine cyclic period of weekly hours. The sine cosine preserve the significant information about a particular week day time and similarity between them. As 23 hours of saturday and 0 hours of sunday would be numerically far away but cyclically they are near

```python
# Combining the Highly correlated features
data['amount_orig'] = data.apply(lambda x: x['oldbalanceOrg'] - x['newbalanceOrig'],axis=1)
data['amount_dest'] = data.apply(lambda x: x['oldbalanceDest'] - x['newbalanceDest'],axis=1)
data['TransactionId'] = data.apply(lambda x: x['nameOrig'] + x['nameDest'],axis=1)
data['hour_of_week'] = (data['step'] % 168)  # Since 1 week = 168 hours

# Convert 'hour_of_week' to sine and cosine values
import numpy as np
data['sin_weekly'] = np.sin(2 * np.pi * data['hour_of_week'] / 168)
data['cos_weekly'] = np.cos(2 * np.pi * data['hour_of_week'] / 168)

#Dropping extra columns
data = data.drop(['oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest','step','nameOrig','nameDest','hour_of_week'],axis=1)

mcolinearTest(data)
```

```python
# Split dataset into training and testing set
from sklearn.model_selection import train_test_split, GridSearchCV
X = data.drop('isFraud', axis=1)
y = data['isFraud']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=22)
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of y_train: ", y_train.shape)
print("Shape of y_test: ", y_test.shape)
```

```python
# import all classifier models for evaluation
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
import time

index = 0 # to track Hypertuned & notfinetuned Classifier based on serial no.
# Best set of tools for model evaluation
def Scores():
    global evaluation, index, X_test, y_test, y_pred
    end = time.time()
    precision, recall, f1, support = precision_recall_fscore_support(y_test, y_pred)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    evaluation = pd.concat([evaluation, pd.DataFrame({
        'index': index,
        'model': model.__class__.__name__,
        'accuracy': model.score(X_test, y_test),
        'precision': [tuple(precision)],
        'recall': [tuple(recall)],
        'f1_score': [tuple(f1)],
        'support': [tuple(support)],
        'True Positive': tp,
        'True negative': tn,
        'False Positive': fp,
        'False Negative': fn,
        'time elapsed': end-start})], ignore_index=True)
    index+=1

    # Visualizing Confusion Matrix
    display = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred.round()))
    display.plot()
    plt.title(f'Confusion Matrix - {model.__class__.__name__}')
    plt.show()

    # Area Under the Receiver Operating Characteristic curve
    # calculate fpr and tpr for all thresholds of the classification
    fpr, tpr, threshold = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(9, 9))
    plt.title(f'ROC {model.__class__.__name__}')
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()

# Metric evaluation dataframe
columns = ['index', 'model', 'accuracy', 'precision', 'recall', 'f1_score', 'support', 'True Positive', 'True negative', 'False Positive', 'False Negative', 'time elapsed']
evaluation = pd.DataFrame(columns=columns)
```

```python
# Random Forest
model = RandomForestClassifier()
start = time.time()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
Scores()
```

```python
# Hyperparameter fine tuning for Random Forest
param_grid = {
    'n_estimators': [9, 22],
    'max_depth': [None, 12]
}
grid_search = GridSearchCV(model, param_grid, cv=2, scoring='f1', refit = True, verbose = 4)
start = time.time()
grid_search.fit(X_train, y_train)
print("Finetuned Random Forest Parameters:", grid_search.best_params_)
```

```python
#  Test Hyperparameter fine tuned Random Forest
model = grid_search.best_estimator_
y_pred = model.predict(X_test)
Scores()
```

```python
# Decision Tree
model = DecisionTreeClassifier()
start = time.time()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
Scores()
```

```python
# Hyperparameter finetuning for Desicion tree
param_grid = {
    'max_depth': [9, 22],
    'criterion': ["gini", "entropy"]
}
grid_search = GridSearchCV(model, param_grid, cv=2, scoring='f1', refit = True, verbose = 4)
start = time.time()
grid_search.fit(X_train, y_train)
print("Finetuned Decision Tree Parameters:", grid_search.best_params_)
```

```python
# Test Hyperparameter Finetuned Decision Tree
model = grid_search.best_estimator_
y_pred = model.predict(X_test)
Scores()
```

The utilization of Random Forest and Decision Tree models is favored over other algorithms due to their ability to handle highly unbalanced datasets (99.87% Genuine, 0.13% Fraud).

The advantage of Random Forest lies in its creation of multiple decision trees, contributing to a nuanced understanding of complex data.
Despite the time-intensive nature of this approach, it proves effective in discerning patterns and capturing fraudulent activity. Also, Decision Trees inherently make decisions in a boolean manner, simplifying the modeling process.

While models like Support Vectior Machine, XGBoost, Bagging, Artificial Neural Networks, and Logistic Regression may yield high accuracy, their limitations in achieving satisfactory precision and recall values make them less suitable for fraud detection applications.

The emphasis on precision and f1 score ensures that innocent individuals are not mistakenly flagged, underscoring the critical importance of accurately identifying fraudulent transactions.
